```python
#import data from csv file
temps_data = pd.read_csv("Average Temperature of Cities.csv")
```

```python
#show dataframe data
temps_data
```

```python
#print number of rows and columns in the dataframe
print(temps_data.shape)
```

```python
#check for null values in data
missing_values_count = temps_data.isnull().sum()
print(missing_values_count[0:16])
```

```python
#drop rows where data is missing
temps_data_dropped = temps_data.dropna()
print(temps_data.shape, temps_data_dropped.shape)
```

```python
#temps_data_filled = temps_data.fillna(method='bfill', axis=0).fillna(0)
#print(temps_data_filled)
#This is not required, as 0 value is a valid temperature value, so do not want to change these
```

```python
#extract all rows for continent value of Europe into separate dataframe
temps_data_europe = temps_data[(temps_data['Continent'] == 'Europe')]
temps_data_europe
```

```python
#create new dataframe containing City and Year columns only
europe_city_year = temps_data_europe[['City','Year']].copy()
print(europe_city_year)
```

```python
#setting index of temps_data_europe dataframe to 'City' column, and removing 'Year' column
temps_data_europe = temps_data_europe.set_index('City')
temps_data_europe = temps_data_europe.drop('Year', axis=1)
temps_data_europe
```

```python
#create function to remove unwanted data from dataframe cells (data in parentheses being removed), and convert values to datatype float
def remove_and_convert(df_name,col_name):
    df_name[col_name] = df_name[col_name].str.replace(r"\(.*\)","", regex=True)
    df_name[col_name] = df_name[col_name].astype(float)
```

```python
#creating list of the column names from temps_data_europe dataframe
month_list = list(temps_data_europe.columns.values)
month_list
```

```python
#removing 'Country' and 'Continent' values from list, as they are not needed
month_list.remove('Country')
month_list.remove('Continent')
month_list
```

```python
#for loop is looping through each month column in temps_data_europe dataframe and running remove_data function to remove data in parentheses
for month_name in month_list :
    remove_and_convert(temps_data_europe,month_name)
```

```python
#check that remove_data function has removed the correct data
temps_data_europe
```

```python
#print datatypes of temps_data_europe columns to check the conversion was successful
print(temps_data_europe.dtypes)
```

```python
#using function to also remove unwanted data and convert datatype in city_year dataframe
remove_and_convert(europe_city_year,'Year')
print(europe_city_year.dtypes)
```

```python
#check function was successful
europe_city_year
```

```python
#group data in temps_data_europe by country, and average temp
countries_grouped = temps_data_europe.groupby(['Country']).mean()
countries_grouped
```

```python
#calculating the yearly average for each country, and appending this as a column to countries_grouped dataframe
count_av = countries_grouped.mean(axis=1)
countries_grouped['Yearly Average'] = count_av
print(countries_grouped.shape)
```

```python
#want to focus on Ireland temps, so apply row with Ireland temps to array
ireland_temps =countries_grouped.iloc[18].to_numpy()
print(ireland_temps)
```

```python
#want to compare UK temps, so apply row with UK temps to array
uk_temps =countries_grouped.iloc[44].to_numpy()
print(uk_temps)
```

```python
#add value of 'Year' to month_list list
month_list.append('Year')
month_list
```

```python
#plot chart of month_list and ireland_temps, to show the average temps in Ireland for each month
x=month_list
y=ireland_temps
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.title('Average temperatures for Ireland')
plt.bar(x,y)
plt.show()
```

```python
#plot chart of ireland temps and UK temps per month, to compare both
x=month_list
y1=ireland_temps
y2=uk_temps
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.title('Average temperatures for Ireland vs Average temperatures for UK')
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
#calaulate average temps per month for all of europe
eur_av = countries_grouped.mean(axis=0)

print(type(eur_av))
print(eur_av)
```

```python
#plot chart of ireland temps, UK temps and europe temps per month, to compare the three
x=month_list
y1=ireland_temps
y2=uk_temps
y3=eur_av.values
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.title('Average temperatures for Ireland and UK vs Average temperatures for Europe')
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```
